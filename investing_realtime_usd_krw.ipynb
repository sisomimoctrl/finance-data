```python
import requests
import pandas as pd
import json
import datetime

headers = {'User-Agent':'Chrome/78.0.3904.87 Safari/537.36',}
url = 'https://api.investing.com/api/financialdata/650/historical/chart/?interval=PT1M&pointscount=70'
json_data = json.loads(requests.get(url, headers=headers).text)
json_data
```

```python
df = pd.DataFrame(json_data['data'])
df.tail()
```

```python
df = df.loc[:,0:5]
df.columns=['Timestamp','Open','High','Low','Close','Volume']
df.tail()
```

```python
df['Timestamp'] = pd.to_datetime(df['Timestamp'], unit='ms')
df.rename(columns = {'Timestamp':'Date'},inplace=True)
df.set_index('Date', inplace=True)
df.tail()
```

```python
# convert timezone from UTC to GMT+9
df.index = df.index.tz_localize('gmt').tz_convert('Asia/Seoul')
df.tail()
```
